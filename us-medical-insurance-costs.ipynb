```python
#Loads the insurance.csv file that will be analyzed
def load_csv(csv_file):
    """
    Loads the insurance.csv file that will be analyzed
    
    Args:
        (str) csv_file: The insurance.csv file of the U.S. Medical Insurance Costs
    Returns:
        df - Pandas DataFrame containing the U.S. Medical Insurance Costs
    """
    #load data file into dataframe
    df = pd.read_csv(csv_file)
    
    
    return df

#Pass the file name to load_csv() function
csv_file = "insurance.csv"
```

Displays the first 5 entries of the top and bottom raw data so that we can see the overview of how the data is structured

```python
#Display the overview of the top and bottom 5 lines of the raw data df
def data_overview(df):
    """
    Displays the to top 5 and bottom 5 information of our raw data
    
    Args:
        (DataFrame) df - Pandas DataFrame containing the U.S. Medical Insurance Costs
    Returns: 
        top_5_data - The first 5 lines of the raw data in the dataframe
        bottom_5_data - The last 5 lines of the raw data in our dataframe     
        
    """
    print("\nDisplay 5 top and bottom lines of the raw data\n")
    
    top_5_data = print("Top 5 lines of the raw data\n", df.head(5))
    bottom_5_data = print("\nBottom 5 lines of the raw data\n", df.tail(5))
    
    return top_5_data, bottom_5_data
```

**Peform Analysis on the following:**
1. General statisticl analysis
2. Average age, Oldest and youngest patients
3. Region where many clients come from 
4. Average, minimum and maximum cost of the insurance

Using **numpy** and **pandas** libraries

```python
#Displays the general statistics of the data
#Prints the Average age, Oldest and youngest patients
#Common regions where people took the medical costs
#Average, minimum and maximum cost of the insurance

def data_analysis(df):
    """
    Computes the General statisticl analysis, Average age, Oldest and youngest patients, Region where many clients come from
    Average, minimum and maximum cost of the insurance
    
    Args:
        (DataFrame) df - dataframe of U.S. Medical Insurance Costs
    Returns: 
        average_cost, maximum_cost, minimum_cost, youngest_person
    """
    #Displays the general descriptive statistics of the data
    print("\n___________Descriptive Statistics of the Data___________\n", df.describe())
    #Computes mean(), max() and min() everage age, oldest and youngest
    average_age =  int(df['age'].mean())
    oldest_person = df['age'].max()
    youngest_person = df['age'].min()
    
    print("\n1. Average Age is {average_age} years".format(average_age = average_age))
    print("2. Oldest person is {oldest_person} years old and the youngest is {youngest_person} years old"
          .format(oldest_person = oldest_person, youngest_person = youngest_person))
    
    #Computes Mode() for region: Where many clients come from
    common_region = df['region'].mode()[0]
    print("3. Most common region: {region}".format(region = common_region))
    
    #Computes mean(), max() and min() for average cost, maximum and minimum insurance costs. Rounded off to 2 decimal places
    average_cost = round(df['charges'].mean(), 2)
    maximum_cost = round(df['charges'].max(), 2)
    minimum_cost = round(df['charges'].min(), 2)
    print("\n4. Analysis on Medical Charges.\n4.1 Average: ${average} \n4.2 Priceciest: ${highest} \n4.3 Cheapest: ${lowest}"
          .format(average = average_cost, highest = maximum_cost, lowest = minimum_cost))
    
    #Highest and average number of children
    average_number_of_children = int(df['children'].mean())
    highest_number_of_children = int(df['children'].max())
    print("\n5. Children Analysis \na) Highest number of children recorded: {max_children} \nb) Average number of children recorded: {ave_children}"
          .format(max_children = highest_number_of_children, ave_children = average_number_of_children))
    
    return average_cost, maximum_cost, minimum_cost, youngest_person
```

**Section B**

Visualize the data using the Matplotlib and Seaborn. Visualize the:
Ages, regions, bmi, charges, smokers and number of children

```python
def visualizing(df):
    """
    Visualize the Ages, regions, bmi, charges, smokers and number of children to gain better understanding
    
    Args:
        (DT) df - dataframe for the U.S. Medical Insurance Costs
        
    Returns:
        Nothing
    """
    
    #Plots the heatmap graph to display the relationship between the gender and smoking
    #cmin = 0.5 means any bin without a count wont be colored, cmap is color map with _r/reverse colores
    print("\n1. Heat Map Showing the relationship between the ages and charges")
    plt.hist2d(data = df, x = df['age'], y = df['charges'], cmin = 0.5, cmap = 'viridis_r');
    plt.colorbar() #default color
    plt.xlabel('Persons Ages') 
    plt.ylabel('Insurance Charges')      
    plt.show()
    
    #Plots the boxplot to display the relationship between costs of smokers and none smokers
    print("\n2. Relationship between the costs of smokers and none smokers")
    base_color = sb.color_palette()[0]
    sb.boxplot(data = df, x = df['smoker'], y = df['charges'], color = base_color);
    plt.xticks(rotation = 10)
    plt.show()
```

```python
def main():
    df = load_csv(csv_file)
    data_overview(df)
    data_analysis(df)
    visualizing(df)

if __name__ == "__main__":
	main()
```

```python
#Create empty lists to be used for storage of data
smoker_status = []
regions = []
charges = []
ages = []
sexes = []
bmi = []
number_children = []
```

```python
#Load the insurance.csv file
def load_data(csv_file, column, lists):
    #Open the csv file
    with open(csv_file) as csv_data:
        #read data from the csv file
        # loops the data and append column names and data into the lists 
        csv_dict = csv.DictReader(csv_data)
        for row in csv_dict:
            lists.append(row[column])
        return lists
#load the data into the lists using load_data
load_data("insurance.csv", "smoker", smoker_status)
load_data("insurance.csv", "region", regions)
load_data("insurance.csv", "charges", charges)
load_data("insurance.csv", "age", ages)
load_data("insurance.csv", "sex", sexes)
load_data("insurance.csv", "bmi", bmi)
load_data("insurance.csv", "children", number_children)
```

```python
#Class to analyze the client's information
class ClientData:
    #Constractor will take the information of all the lists
    def __init__(self, smoker_status, regions, charges, ages, sexes, bmi, number_children):
        self.smoker_status = smoker_status
        self.regions = regions
        self.charges = charges
        self.ages = ages
        self.sexes = sexes
        self.bmi = bmi
        self.number_children = number_children
    #Method to analyze ages in the dataset
    def ages_analysis(self):
            #initialize total_ages, youngest_client, oldest_client
            total_ages = 0 
            youngest_client = 200
            oldest_client = 0
            #Loop through the ages list
            for age in self.ages:
                age = int(age)
                total_ages += int(age) #Sum ages
                #find the oldest and youngest client
                if age > oldest_client:
                    oldest_client = age
                if age < youngest_client:
                    youngest_client = age
            #calculate the average age
            average_age = round(total_ages/len(self.ages), 0)
            print("Oldest client is: {oldest} and Youngest client is: {youngest}"
                  .format(oldest = oldest_client, youngest = youngest_client))
            return  "Average age is: {age}".format(age = average_age)
    
    #Analyze average charges on clients        
    def charges_analysis(self):
        #initialize total_charges, highest_charge, lowest_charge to 0, 0 and first element of charges
        total_charges, highest_charge, lowest_charge = 0, 0, float(self.charges[0])
        #Loop through the charges list
        for charge in self.charges:
            charge = float(charge)
            total_charges += charge #sum up charges
            #find the highest insurance cost
            if charge > highest_charge:
                highest_charge = charge
            #find the lowest insurance cost
            if charge < lowest_charge:
                lowest_charge = charge
        #compute average charge
        average_charges = round(total_charges/len(self.charges), 2)
        
        print("\nLowest insurance cost is : ${lowest} and highest insurance cost is: ${high}"
              .format(lowest = lowest_charge, high = highest_charge))
        return "Average charges on clients: ${charges}".format(charges = average_charges)
        
    #Records clients information into a dictionary
    def build_dictionary(self):
        #initialize an empty dictionary
        self.clients_dictionary = {}
        self.clients_dictionary['age'] = [int(age) for age in self.ages]
        self.clients_dictionary['sex'] = self.sexes
        self.clients_dictionary['bmi'] = self.bmi
        self.clients_dictionary['children'] = self.number_children
        self.clients_dictionary['smoker'] = self.smoker_status
        self.clients_dictionary['region'] = self.regions
        self.clients_dictionary['charges'] = [float(charge) for charge in self.charges]
        
        return self.clients_dictionary
    
    #Performs analysis using the clients dictionary
    def client_dictionary_analysis(self):
        #initialize empty list of regions
        self.unique_regions = []
        #Iterate through the dictionary
        self.build_dictionary()
        #access regions in the dictionary and append regions in the list
        for region in self.clients_dictionary['region']:
            if region not in self.unique_regions:
                self.unique_regions.append(region)
        #count how many clients come from each region
        southwest_count = self.clients_dictionary['region'].count('southwest')
        southeast_count = self.clients_dictionary['region'].count('southeast')
        northwest_count = self.clients_dictionary['region'].count('northwest')
        northeast_count = self.clients_dictionary['region'].count('northeast')
        
        print("\nClient are from Regions: {}".format(self.unique_regions))
        print("\nClients are distributed as follows:")
        print("a) South West: {sw} \nb) South East: {se} \nc) North West: {nw} \nd) North East: {ne}"
             .format(sw = southwest_count, se = southeast_count, nw = northwest_count, ne = northeast_count))
        
        #count number of males, females and number of smokers and nonsmokers using attributes
        males_count = self.sexes.count('male')
        females_count = self.sexes.count('female')
        smokers_count = self.smoker_status.count('yes')
        non_smoker_count = self.smoker_status.count('no')
        print("\nNumber of Males and Females \na) Males: {m_count} \nb) Females: {f_count}"
              .format(m_count = males_count, f_count = females_count))
        print("\nNumber of smokers and Non-smokers \na)Smokers: {smoker} \nb) Non-smokers: {nonsmoker}"
              .format(smoker = smokers_count, nonsmoker = non_smoker_count))
        
        return males_count, females_count, smokers_count, non_smoker_count
    
    def charges_analysis(self):
        #zip lists using attributes
        #initialize cost_1 to 0 and cost_2 to the first element of charges list
        cost_1, cost_2 = 0, float(self.charges[0])
        #itereate through the zip(lists)
        for age,charges, bmi, smoker_status, number_children in zip(self.ages, self.charges, self.bmi, self.smoker_status, self.number_children):
            #charges are explicitly converted into floating point numbers
            #Analysis of highest and lowest charges
            charges = float(charges)
            if cost_1 < charges:
                cost_1 = round(charges, 2)
                age_1, bmi_1, smoker_1, number_children_1 = int(age), bmi, smoker_status, number_children
            
            if cost_2 > charges:
                cost_2 = round(charges, 2)
                age_2 = age
                bmi_2 = bmi
                number_children_2 = number_children
                smoker_2 = smoker_status
                
        print("\nClient who is {years} years old, with bmi of {bmi} and {number_children} child(ren) is associated with the insurance cost of ${cost}"
              .format(years = age_1, cost = cost_1, bmi = bmi_1, number_children = number_children))
        print("Does the client smoke?: {smoke}".format(smoke = smoker_1))
        
        print("Client who is {years} years old, with bmi of {bmi} and {number_children} child(ren) is associated with the insurance cost of ${cost}"
              .format(years = age_2, cost = cost_2, bmi = bmi_2, number_children = number_children_2))
        print("Does the client smoke?: {smoke}".format(smoke = smoker_2))
        
        #Analysis of charges in respect to bmi and number of children using attributes
        bmi_1, bmi_2 = self.bmi[0], self.bmi[0] #initiaize bmi to the first element of bmi
        for bmi in self.bmi:
            #find the lowest bmi
            if float(bmi_1) > float(bmi):
                bmi_1 = bmi
            #find the highest bmi
            if float(bmi_2) < float(bmi):
                bmi_2 = bmi                
        
        print("\nThe lowest bmi is: {bmi}".format(bmi = bmi_1))
        print("The highest bmi is: {bmi}".format(bmi = bmi_2))
        
        return bmi_1, bmi_2
```

```python
client_info = ClientData(smoker_status, regions, charges, ages, sexes, bmi, number_children)

print(client_info.ages_analysis())
print(client_info.client_dictionary_analysis())
print(client_info.charges_analysis())
```
